```python
# Import necessary libraries #
from utils import dates
from utils import gantt_chart_functions
```

```python
# Importing Gant Chart Data #
gantt_chart_data = gantt_chart_functions.get_gantt_chart_data()
gantt_chart_data
```

```python
# Converting Dates from Strings into Datetimes and Arranging Dataframe for Gantt Chart #
gantt_chart_data = dates.separate_dates(gantt_chart_data)
gantt_chart_data = dates.calculate_time_differences(gantt_chart_data)
gantt_chart_data = gantt_chart_data.sort_values(by="Start_Date", ignore_index=True)
gantt_chart_data
```

```python
# Colour Assigning #
gantt_chart_data_colour_conditioned = gantt_chart_functions.assign_conditional_colours(gantt_chart_data)
gantt_chart_data_colour_conditioned
```

```python
# Displaying Gantt Chart #
gantt_chart_functions.display_gantt_chart(gantt_chart_data_colour_conditioned, "Commercial")
```
